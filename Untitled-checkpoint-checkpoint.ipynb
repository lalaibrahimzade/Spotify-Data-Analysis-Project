```python
most_popular=df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python

df_tracks.rename(columns=lambda x: x.strip(), inplace=True)
print(df_tracks.columns)
print(df_tracks.head())

# Reset the index to convert 'release_date' back to a column
df_tracks.reset_index(inplace=True)

df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce')
df_tracks.set_index("release_date", inplace=True)

# Inspect the head of the dataframe
print(df_tracks.head())
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
# Inspect a few rows to understand the structure
print(df_tracks.head())

# If the column is present, convert duration and drop duration_ms
if 'duration_ms' in df_tracks.columns:
    df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x / 1000))
    df_tracks.drop("duration_ms", inplace=True, axis=1)
    print(df_tracks.head())
else:
    print("Column 'duration_ms' not found in the DataFrame.")
```

```python
df_tracks.duration.head()
```

```python
df_numeric = df_tracks.select_dtypes(include='number').drop(["key", "mode", "explicit"], axis=1)
corr_df = df_numeric.corr(method="pearson")
plt.figure(figsize=(18, 10))
heatmap = sns.heatmap(
    corr_df, 
    annot=True, 
    fmt=".1g", 
    vmin=-1, 
    center=0, 
    cmap="inferno", 
    linewidths=0.5, 
    linecolor="black",
    annot_kws={"size": 10}
)
heatmap.set_title("Correlation Heatmap Between Variables", fontsize=16)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, ha='right', fontsize=10)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=10)
plt.tight_layout()
plt.show()
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x=years, y=total_dr, ax=ax, err_kws={'linewidth': False}).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre = pd.DataFrame({
    'genre': ['Rock', 'Pop', 'Jazz', 'Classical', 'Hip-Hop', 'Electronic', 'Reggae', 'Country', 'Blues', 'Metal'],
    'popularity': [80, 70, 60, 50, 90, 65, 55, 45, 75, 85]  # Replace with actual popularity data
})

# Check if 'popularity' column exists
if 'popularity' not in df_genre.columns:
    raise KeyError("'popularity' column not found in DataFrame")

# Set Seaborn style
sns.set_style("darkgrid")

# Create a figure with a specific size
plt.figure(figsize=(10, 5))

# Sort genres by popularity and get top 10
famous = df_genre.sort_values("popularity", ascending=False).head(10)

# Create a horizontal bar plot
sns.barplot(y='genre', x='popularity', data=famous).set(title="Top 10 Genres by Popularity")

# Show plot with labels
plt.xlabel("Popularity")
plt.ylabel("Genre")
plt.show()
```
